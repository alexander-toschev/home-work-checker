# 🧮 NumPy Assignment (Colab-ready) — ввод ФИО/группы/ID + автопроверка

**Как работать (в Colab или JupyterLab):**
1. Введите данные в блоке «Данные студента» (имя, фамилия, группа, ID задания).
2. Решайте задачи в блоках **TODO** — меняйте код в отмеченных местах.
3. В конце запустите ячейку **Автопроверка** — она посчитает баллы и **выведет** JSON-строкой результат.
   - Последняя строка последней ячейки будет вида: `{"name": "...", "group": "...", "assignment": "...", "score": N}`
   - Это совместимо с нашим проверяющим хранилищем.

**Баллы:** 100 максимум.
